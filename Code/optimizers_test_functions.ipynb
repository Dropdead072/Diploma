```python
import numpy as np
from utils.objective_functions.test_functions import * 
from utils.optimizers.numpy_optimizers import *
from core.optimizer import optimize
```

```python
def random_init(obj: "ObjectiveFunction", seed=None):
    """
    Генерирует случайную стартовую точку внутри bounds объекта.
    
    Args:
        obj: экземпляр ObjectiveFunction
        seed: опционально, фиксируем для воспроизводимости
        
    Returns:
        np.ndarray с shape (dim,)
    """
    rng = np.random.default_rng(seed)
    x0 = np.array([rng.uniform(low, high) for (low, high) in obj.bounds])
    return x0
```

```python
func = Rosenbrock()
opt = GradientDescent(lr=1e-5)

x0 = random_init(func, 4)

x_star, log = optimize(func, opt, x0, max_iter=10**5)

print("Converged to:", x_star)
print("Final value:", log[-1]["f"])
```
